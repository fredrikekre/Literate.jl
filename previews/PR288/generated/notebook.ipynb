# Rational numbers

In julia rational numbers can be constructed with the `//` operator.
Lets define two rational numbers, `x` and `y`:

```julia
x = 1 // 3
```

```julia
y = 2 // 5
```

When adding `x` and `y` together we obtain a new rational number:

```julia
z = x + y
```
